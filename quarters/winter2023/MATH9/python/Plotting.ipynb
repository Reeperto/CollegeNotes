Example: plot cos functions

```python
fig, ax = plt.subplots()
x = range(7)
ax.plot(x,np.cos(x))
```

```python
fig, ax = plt.subplots()
x = np.linspace(0,7,100)
ax.plot(x,np.cos(x))
```

```python
# another approach is to use plt.plot
fig, ax = plt.subplots()
x = np.linspace(0,7,100)
plt.plot(x,np.cos(x))
```

## Multiple plots in one or serveral figures

```python
fig, ax = plt.subplots()
x = np.linspace(0,7,100)
ax.plot(x,np.cos(x))
ax.plot(x,np.cos(2*x))
```

Example: plot cos functions with frequence 1,2,3,4,5,6

```python
k_list = [1,2,4,4,5,6]
```

```python
fig, ax = plt.subplots()
for k in k_list:
    f = lambda x: np.cos(k*x)
    y = f(x)
    ax.plot(x,y)
```

```python
fig, axs = plt.subplots(3,2)
axs_flat = axs.reshape(-1) # enable us to call the index of each figure

for i in range(6):
    k = k_list[i]
    axs = axs_flat[i]
    f = lambda x: np.cos(k*x)
    y  = f(x)
    axs.plot(x,y)
```
